```python
for i in range(len(earthquake_info)) : 
    earthquake_info.loc[i, '시간'] = earthquake_info.loc[i, '시간'].split(' ')[0]
earthquake_info
```

```python
earthquake_info.drop(['위도', '경도'], axis = 1, inplace = True) # 필요없는 열 제거

earthquake_info['진앙(km)'].replace(r'^\s*$', np.nan, regex=True, inplace=True) # 공백값 null로 대체


check = earthquake_info['진앙(km)'].isna() # null값인 곳 check

earthquake_info
```

```python
average = 0
count = 0

for i in range(len(earthquake_info)):
    if(not check[i]) : # null값 확인 (true false)
        count = count + 1
        average = average + int(earthquake_info.loc[i]['진앙(km)'])
    
    else : 
        print(i)

if(count > 0) : 
    average = average / count # 결측값 제외 모든 값을에 대해 평균을 구함
    
earthquake_info['진앙(km)'] = earthquake_info['진앙(km)'].fillna(average) # 결측값을 가지고 있는 해당 값을 평균값으로 대체

earthquake_info['진앙(km)'] = pd.to_numeric(earthquake_info['진앙(km)']) 

earthquake_info_data = earthquake_info.groupby('시간').mean()

earthquake_info_data.rename(columns={'진앙(km)':'진앙'}, inplace=True)
```

```python
stock_mean = quakestock_1.copy()

i = 1

for i in range(len(stock_mean)) : 
    stock_mean.loc[i, 'sise'] += quakestock_2.loc[i, 'sise'] + quakestock_3.loc[i, 'sise']
    stock_mean.loc[i, 'sise'] = int(stock_mean.loc[i, 'sise'] / 3)
    stock_mean.loc[i, 'trading'] += quakestock_2.loc[i, 'trading'] + quakestock_3.loc[i, 'trading']
    stock_mean.loc[i, 'trading'] = int(stock_mean.loc[i, 'trading'] / 3)
stock_mean.date = pd.to_datetime(stock_mean.date)
stock_mean
```

```python
earthequake_power = earthquake_info_data.copy()
earthequake_power['power'] = 0
for i in earthequake_power.index : 
    if earthequake_power.loc[i, '진앙'] == 0 : 
        earthequake_power.loc[i, '진앙'] =  1
    earthequake_power.loc[i, 'power'] = earthequake_power.loc[i, '규모'] /earthequake_power.loc[i, '진앙']

earthequake_power = earthequake_power.reset_index()  # 날짜를 뺀다.
earthequake_power.columns = ['시간', '규모', '진앙', 'power']

earthequake_power
```

```python
earthquake_check = (earthequake_power['시간'] >= '2017-02-01') & (earthequake_power['시간'] <= '2017-04-17')
earthquake_check_data = earthequake_power.loc[earthquake_check]
earthquake_check_data = earthquake_check_data.reset_index(drop = True)
earthquake_check_data = earthquake_check_data.astype({'시간' : 'datetime64[ns]'})

index = pd.date_range('2017-02-01', '2017-04-22')

df = earthquake_check_data.set_index('시간')
df2 = df.reindex(index, fill_value=np.nan)

df3 = df2.interpolate()

average_size = df3['규모'].mean()
average_fd = df3['진앙'].mean()
average_power = df3['power'].mean()

for i in range(4) :
    df3.iloc[i, 0] = average_size
    df3.iloc[i, 1] = average_fd
    df3.iloc[i, 2] = average_power

earthquake_check_data = df3

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import rc, font_manager

rc('font', family = "Segoe UI")

fig, axis1 = plt.subplots()
axis1.plot(earthquake_check_data.index, earthquake_check_data.규모, color='red', label='size', alpha=0.3)

axis2 = axis1.twinx()
axis2.plot(earthquake_check_data.index, earthquake_check_data.진앙, color='blue', label = 'focal distance')


fig.set_size_inches(10, 10)
fig.tight_layout()
fig.legend()
plt.show()
```
